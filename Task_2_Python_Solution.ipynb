##**Section 1: Lists**

####**1. Create a List:**
Create a list containing the numbers 1 through 15. Print the list

```python
# 1. Using for-loop
lst = []

for i in range(1, 16):
  lst.append(i)

# 2. Using list comprehension
num_lst = [i for i in range(1, 16)]

# Printing out the result
print("1. Using for-loop: \n\tList = ", lst)
print("\n2. Using List Comprehension: \n\tList = ", num_lst)

```

> **Note:**
*Python List Comprehension* **bold text**provides a concise syntax for creating new lists based on existing iterables (like lists, strings, or ranges). It is often more readable and faster than traditional for loops.

> **Syntax:**
*new_list = [expression for item in iterable if condition]*
