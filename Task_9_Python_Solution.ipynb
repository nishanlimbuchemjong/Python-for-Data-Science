- Write a Python function using def to return the product of any count of numbers that is passed as an argument to the function.

```python
# Creating a function
def product(*num):
  mul = 1
  for i in num:
    mul = mul * i
  return mul

# Calling a function
result = product(5, 4)
print("Total product = ", result)
```

-	Use reduce() from functools to calculate the product of a list [1, 2, 3, 4, 5].

```python
from functools import reduce

def product(number):
  return reduce(lambda x, y: x * y, number )

print("Total product: ", product([1, 2, 3, 4, 5]))
```

- Write a function that takes different objects (Rectangle, Circle, Square) and calls their area() method.

```python
class Rectangle:
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth

  def area(self):
    return self.length * self.breadth

class Square:
  def __init__(self, length):
    self.length = length

  def area(self):
    return self.length ** 2

class Circle:
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return (22 * (self.radius) ** 2) / 7

rec = Rectangle(4, 5)
sqr = Square(4)
cir = Circle(2)


print("Area = ", cir.area())
print("Area = ", rec.area())
print("Area = ", sqr.area())
```

-	Implement a class Calculator with methods for addition, subtraction, multiplication, and division.

```python
class Calculator:
  def addition(self, *num):
    return sum(num)

  def subtraction(self, *num):
    x = num[0]
    for i in num[1:]:
      x -= i
    return x

  def multiplication(self, *num):
    x = 1
    for i in num:
      x *= i
    return x

  def division(self, *num):
    x = num[0]
    for i in num[1:]:
      if num == 0:
        return 'Cannot be divided by zero.'
      else:
        x /= i
    return x

cal = Calculator()
print("Addition = ", cal.addition(4, 4, 40))
print("Subtraction =", cal.subtraction(40, 4, 4))
print("Multiplication = ", cal.multiplication(4, 4, 4))
print("Division = ", cal.division(40, 2, 5))
```

-	Use the ABC module to define an abstract class Shape with an abstract method calculate_area() and implement it in Circle and Rectangle subclasses.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def calculate_area(self):
    return (22 * (self.radius **2)) / 7

class Rectangle(Shape):
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth

  def calculate_area(self):
    return self.length * self.breadth

cir = Circle(4)
rec = Rectangle(4, 2)

print("Area of Circle = ", cir.calculate_area())
print("Area of Rectangle = ", rec.calculate_area())
```
