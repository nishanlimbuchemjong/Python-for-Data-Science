####**Equality vs. Identity Check (Conceptual Explanation)**

> **Equality Operator (==) :**
used when we want to check whether two objects contain same data (value), regardless of whether they are stored in same memory location. Internally, it calls ___eq_ __() method of class. For built-in types like lists, strings and numbers, this means checking if the contents are same.



```python
x = [1, 2, 3]
y = [1, 2, 3]

if x == y:
  print("x and y have same value")
else:
  print("x and y doesn't have same value")

"""
x and y are two different variables (objects)
but both contains same value (56), x == y returns True
(==) only looks at values, not memeory addresses
"""
```

> **Identity Operator (Is) :**
used when we want to check whether two variables refer to exact same object in memory. It does not care about values. Even if two objects look identical, it will return False unless both variables literally point to the same memory location.

```python
x = [4, 5, 6]
y = [4, 5, 6]
z = x

if x is y:
  print("x and y are the same object")
else:
  print("x and y are not the same object.")

if x is z:
  print("x and z are the same object")
else:
  print("x and z are not the same object.")

"""
x and y have same value but the location are different. so (x is y) returns False

x and z have same value and also same locatin. so (x is z) returns True.
"""
```
