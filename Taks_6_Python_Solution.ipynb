####**Print Numbers from 10 to 30**
- Use a while loop to print numbers from 10 to 30.

```python
# Using while-loop to print nubmers from 10 to 30
a = 10
while a <= 30:
  print(a)
  a += 1
```

####**Print Odd Numbers from 1 to 20**
- Use a while loop to print all odd numbers from 1 to 20.

```python
# Using while loop to print odd numbers from 1 to 20
a = 1

print("Odd number from 1 to 20:")
while a <= 20:
  if a % 2 != 0:
    print(a)
  a += 1
```

####**Print a Word 5 Times**
- Use a while loop to print the word "Hello" five times.

```python
# Using while loop to print 'Hello' five times
a = 1
while a <= 5:
  print("Hello")
  a += 1
```

####**Countdown From 5**
- Use a while loop to print numbers from 5 to 1.

```python
# Using while loop to print nubmers from 5 to 1
a = 5

print("Counting.....")
while a >= 1:
  print(a)
  a -= 1
```

####**Print All Multiples of 3 upto 30**
- Use a while loop to print 3, 6, 9, ..., up to 30.

```python
# Using while loop to print Multiples of 3 upto 30
a = 1

print("Multiple of 3:")
while a <= 10:
  print(f"3 X {a} = {3 * a}")
  a += 1
```

####**Keep Asking for a Number Unitl It's Positive**
- Keep asking the user to enter a number until they enter a positive number.

```python
# Asking the user to enter a number until they enter a positive number

while 1:
  num = int(input("Enter a number: "))
  if num < 0:
    print("\nIt is negative number.\nExiting.....")
    break
```

####**Guess the Secret Number using While Loop**
- Secret number = 7.

```python
# Using while loop to guess the secret number
secret_num = 7

while 1:
  guess = int(input("Enter a number: "))
  if guess ==secret_num:
    print("Congratulation, you guess it right")
    break
```

####**Function to Add Two Numbers**
- Create a function add(a, b) that returns the sum.

```python
# Createing a function for addition
def addition(x, y):
  '''
  It takes two parameters and return total sum
  '''
  return x + y

# Calling an addition() function
a, b = 5, 5
print(f"Addition of {a} and {b} = {addition(a, b)}")
```

####**Function to Multiply Two Numbers**
- Create a function multiply(x, y) that returns the product.

```python
# Creating a function for multiplying two numbers
def multiply(x, y):
  '''
  this function takes two parameters and return its product
  '''
  return x * y

# Calling a function
print("Multiplication = ", multiply(5, 5))
```

####**Function to Check Even or Odd**
- Create a function check_even(num) that prints "Even" or "Odd".

```python
# Defining a functionn to check even or odd
def check_even(num):
  '''
  this function takes one parameters, checks even or odd then reutrn the string
  '''
  if num % 2 == 0:
    return "It is Even"
  else:
    return "It is Odd"

# Calling a function
print(check_even(56662341))
```
