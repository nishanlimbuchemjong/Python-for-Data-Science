####**1. Encapsulation**
- Create a class Person with private attributes __name and __age.
- Use getter and setter methods to access and modify these values.


```python
class Person:
  def __init__(self, name, age):
    self.__name = name
    self.__age = age

  def get_name(self):
    print(f"Name = {self.__name}")

  def get_age(self):
    print(f"Age = {self.__age}")

  def set_name_age(self, name, age):
    self.__name = name
    self.__age = age

per_obj = Person('Nishan', 26)
per_obj.get_name()
per_obj.get_age()

print("\nAfter Updating:")
per_obj.set_name_age('Limbu', 25)
per_obj.get_name()
per_obj.get_age()
```

####**2. Polymorphism**
- Create a function show_details(obj) that works with different classes.
- Define the same method name (details()) in two or more classes.
- Call show_details() with different objects to show polymorphism.


```python
class Teacher:
  def __init__(self, name, subject):
    self.name = name
    self.subject = subject

  def details(self):
    print(f"Teacher:\n\tName: {self.name}\n\tSubject: {self.subject}")

class Student:
  def __init__(self, name, course):
    self.name = name
    self.course = course

  def details(self):
    print(f"Student:\n\tName: {self.name}\n\tCourse: {self.course}")


# Creating a function
def show_details(obj):
  obj.details()

# Passing datas
teacher = Teacher('Rakesh Bachchan', 'AI/ML')
student = Student('Nishan Limbu', 'AI/ML Engineer')

show_details(teacher)
show_details(student)
```

####**5. Hierarchical Inheritance**

####**6. Polymorphism Using a Loop**
- Create classes Pen and Pencil, both with a method write().
- Loop through a list of objects (Pen and Pencil) and call write() on each object.


```python
class Pen:
  def write(self):
    print("Writing by Pen")

class Pencil:
  def write(self):
    print("Writing by Pencil")

lst =[Pen(), Pencil(), Pencil(), Pencil(), Pen(), Pen(), Pencil(), Pen()]

# Looping through a list of objects
for obj in lst:
  obj.write()

```

####**7. Use Encapsulation to Protect Age**
- Create a class Person with private attribute __age.
- Use set_age() to set the age.
- If age is less than 0, print "Invalid age".
- Use get_age() to return the age.


```python
class Person:
  def __init__(self, age):
    self._age = age

  def get_age(self):
    return self._age

  def set_age(self, age):
    if age > 0:
      self._age = age
    else:
      print("Invalid age")

per_obj = Person(25)
print("Age = ", per_obj.get_age())

print("\nAfter Updating Age:")
per_obj.set_age(26)
print("Age= ", per_obj.get_age())
```

####**8. Company Structure**
- Create class Company with method company_name().
- Create class Department that inherits Company and adds department_name().
- Create class Employee that inherits Department and adds employee_name().
- Create an Employee object and call all the methods.


```python
class Company:
  def __init__(self, companyname):
    self.companyname = companyname

  def company_name(self):
    print(f"Company name = {self.companyname}")

class Department(Company):
  def __init__(self, companyname, departmentname):
    super().__init__(companyname)
    self.departmentname = departmentname

  def department_name(self):
    print(f"Department name = {self.departmentname}")

class Employee(Department):
  def __init__(self, companyname, departmentname, employeename):
    super().__init__(companyname, departmentname)
    self.employeename = employeename

  def employee_name(self):
    print(f"Employee name = {self.employeename}")

emp_obj = Employee('ABC', 'Administration', 'Nishan Limbu')
emp_obj.company_name()
emp_obj.department_name()
emp_obj.employee_name()
```

####**9. Lambda to check if a Number is Positive**
- Write a lambda function that takes a number and returns True if it is positive, otherwise False.
- Test it with different values like 5, -3, and 0


```python
check_positive = lambda x: 'True' if x > 0 else 'False'

print(check_positive(5))
print(check_positive(-3))
print(check_positive(0))
```

####**11. Add 10 to All List Elements**
- Given a list [1, 2, 3, 4, 5],
- Use map() and lambda to return a new list where 10 is added to each number.


```python
lst = [1, 2, 3, 4, 5]

# Using map() and lambda to return a new list where 10 is added to each number
new_lst = list(map(lambda x: x + 10, lst))

print("Old list = ", lst)
print("New List = ", new_lst)
```

####**12. Get Length of Each Word**

- Use map() and lambda to return a list of lengths of each word.
- Output: [5, 6, 4]

####**13. Filter Numbers Greater Than 50**
- Given a list [10, 55, 60, 23, 90],
- Use filter() and lambda to return only the numbers greater than 50.


```python
lst = [10, 55, 60, 23, 90]

# Using filter() and lambda to return only the nubmers greater than 50
greater_number_than_50 = list(filter(lambda x: x > 50, lst))
print(greater_number_than_50)
```

####**14.Create a Word by Joining Letters**
- Given a list of letters ['P', 'Y', 'T', 'H', 'O', 'N'],
- Use reduce() and lambda to join them into a single word: "PYTHON".


```python
from functools import reduce

letters = ['P', 'Y', 'T', 'H', 'O', 'N']

# Using reduce() and lambda to join into a single word
word = reduce(lambda x, y:"".join(letters), letters)

# printing the result
print('letters = ', letters)
print("After joining:")
print(f"Single word = {word}")
```
