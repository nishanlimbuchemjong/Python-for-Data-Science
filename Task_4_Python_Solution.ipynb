###**Grade Categorizer**

**Task:** Prompt the user for a number between 0 and 100 (the score).
- Use if/elif/else to categorize the score:
 - 90–100: "Grade A"
- 80–89: "Grade B"
- 70–79: "Grade C"
- < 70: "Fail"


```python
# Asking a user for a number between 0 and 100
num =int(input("Enter a number between 0 and 100: "))

# Checking for the user score
if num>100 or num < 0:
  print("Warning: Number should be between 0 and 100")
else:
  if num > 90:
    print("Grade A")
  elif num > 80:
    print("Grade B")
  elif num > 70:
    print("Grade C")
  else:
    print("Fail")
```

###**Count Vowels in a String**

**Task:** Ask the user for a string.
- Use a for loop to iterate over each character.
- Use if conditions to check if it’s a vowel ("a", "e", "i", "o", "u").
- Count the total number of vowels and print the result.


```python
# Initialize a variable
count = 0
vowel = ['a', 'e', 'i', 'o', 'u']
# Ask users to enter a string
word = input("Enter  a string: ")

# Counting vowels in a string
for char in word:
  if char in vowel:
    count = count + 1

# Printing the result
print("You typed = ", word)
print("Total number of vowels = ", count)
```

###**Categorize Scores**

**Task:** Have a list of scores, e.g., scores = [85, 42, 78, 99, 65].
- Use a for loop to iterate over each score.
- Use if/elif/else inside the loop to categorize each score as "Pass" (>= 60) or "Fail" (< 60).
- Print each score with its category.


```python
# Initialize a list
scores = [85, 42, 78, 99, 65]

# checking pass or fail from the score list using for-loop
for i in scores:
  if i >= 60:
    print(f"{i} => (Pass)")
  else:
    print(f"{i} => (Fail)")
```
