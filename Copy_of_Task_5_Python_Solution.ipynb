####**Iterate Through a Tuple and Print Types**
**Task:** Create a tuple with at least 5 different types of elements (int, float, string, bool, and complex).
- Use a for loop to iterate over the tuple and print each element along with its data type.


```python
# Creating a Tuple
tup = (34, 465, True, 345.43, False, 'Nishan', 3 - 3j, 2343.3333, 'Limbu')

# Using For-Loop for iteration over the tuple
for item in tup:
  print(f"Item = {item}, Data type = {type(item)}")
```

####**Print Each Character of a String with Its Index**
**Task:** Ask the user to enter a word.
- Use a for loop and enumerate() to print each character of the string along with its index.


```python
# Asking a user to enter a word
word = input("Enter a word: ")

# Using a for-loop and enumerate() to print each character of the string along with its index
for index, char in enumerate(word, start=1):
  print(f"{index} = {char}")
```

####**Sum of Elements in a List**
**Task:** Create a list of integers.
-	Use a for loop to calculate the sum of all the elements and print the total.


```python
# Creating a list of integers
lst =[1, 2, 3, 4, 345, 2342, 45346, 54675756]
sum = 0
# Using for-loop to calculate the sum of all element
for i in lst:
  sum += i

# Printing the output
print("List = ", lst)
print("Sum = ", sum)
```

####**Count Booleans in  a Tuple**
**Task:** Create a tuple containing different data types, including multiple True and False values.
- Use a for loop to count how many True values are present and print the count.


```python
# Creating a tuple containing different data types
tup = (True, True, True, 34, 465, True, 345.43, False, 'Nishan', 3 - 3j, 2343.3333, False, False, False, False, False, 'Limbu')
count = 0

# Using for-loop to count the number of True values present in tuple
for i in tup:
  if i == True:
    count += i

# Printing out the result
print("Tuple = ", tup)
print("Total count of True value = ", count)
```

####**Check for Vowels in a String**
**Task:** Ask the user for a word.
- Use a for loop to check each character and print a message if it’s a vowel (a, e, i, o, u).


```python
# Initialize a list
vowel = ['a', 'e', 'i', 'o', 'u']

# Asking a user to enter a word
word = input("Enter a word: ")

# Using for-loop to check each character if it is a vowel or not
for char in word:
  print(f"{char} --> {"Yes Vowel" if char in vowel else "No Vowel"}")
```

####**Print Square of Numbers from a Tuple**
**Task:** Create a tuple of numbers from 1 to 5.
- Use a for loop to iterate through the tuple and print the square of each number.


```python
# Creating a tuple fo numbers from 1 to 5 using for-loop
lst = []

for i in range(1, 6):
  lst.append(i)

tup = tuple(lst) # tuple created after converting list into tuple
print("Tuple = ", tup)

# Using for-loop to iterate through the tuple and printing square of each number
for i in tup:
  print(f"Number = {i}, Square = {i * i}")
```

####**Print Elements of a List in Uppercase**

```python
# Creating a list of 5 small letter containing words
lst = ['nisha', 'limbu', 'kavin', 'sorry', 'apple']

# Using for-loop to iterate and printing each word in uppercase
for ele in lst:
  print(f"word = {ele}, uppercase = {ele.upper()}")
```

####**Count Numbers Greater Than 10 in a List**
**Task:** Create a list of integers.
- Use a for loop to count how many numbers in the list are greater than 10.
- Print the final count.


```python
# Creating a list of integers
lst = [1, 2, 3, 4, 44, 64, 323, 3454, 6676]
count = 0

# Using a for-loop to count numbers greater than 10
for num in lst:
  if num > 10:
    count += 1

# Printing out the result
print("List = ", lst)
print("Total number greater than 10 = ", count)
```
