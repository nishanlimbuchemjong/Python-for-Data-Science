###**Casefold vs Lower**

-	Create two strings, "Case" and "case".
-	Compare them with the regular == operator directly.
-	Compare them again after applying .casefold().
-	Print results and comment on how .casefold() differs from .lower() in edge cases.


```python
# Creating two strings
s1, s2 = 'Case', 'case'

print("Strings:\n\tString-1 = {}\n\tString-2 = {}".format(s1, s2))

# Comparing them using (==) operator
print("\nAfter comparing them using (==), \n\t = ", s1 == s2)

# Comparing them again after applying .casefold()
print("\nComparing after applying .casefold() , \n\t = ", s1.casefold() == s2.casefold())
```

> **Note:**
- ***casefold() method*** in python returns a copy of a string where all characters are converted to a standardized lowercase form.
- ***lower() method*** is used to display or format the text. It works predictably for the ASCII character set but may not handle certain special or international characters correctly for comparison.

```python
# Given list
fruits = ["apple", "banana", "cherry"]

print("Original list\n\tFruits = ", fruits)

# Sorting them in descending alphabetical order
desc_fruits = fruits[::-1]
print("\nSorting them in descending alphabetical order,")
print("\tFruits = ", desc_fruits)

# Reversing the sorted list using .reverse()method to flip it
desc_fruits.reverse()
print("\nReversing the sorted list,")
print("\tFruits = ", desc_fruits)

# Compare the both lists
print("\nComparing lists:")
print("\tIs original fruits == reversed fruits? ", fruits == desc_fruits)
```

###**Tuple Concatenation & Replication**

-	Create two tuples (1, 2) and (3, 4).
-	Concatenate them into (1, 2, 3, 4).
-	Replicate (1, 2) 3 times to get (1, 2, 1, 2, 1, 2).


```python
# Creating two tuples
t1, t2 = (1, 2), (3, 4)

print("Original tuples:\n\tTuple-1 = {}\n\tTuple-2 = {}".format(t1, t2))

# Concatenation
print("\nAfter Concatenation,\n\tTuple = {}".format(t1+t2))

# Replicate (1, 3), three times
print("\nAfter replicating (1, 3) three times, we get\n\tTuple-1 = {}".format(t1*3))
```

###**Intersection & Union**

-	Let setA = {1, 2, 3, 4} and setB = {1,2,3}.
-	Print their intersection using setA & setB.
-	Print their union using setA | setB.


```python
# Given variable
setA, setB = {1, 2, 3, 4}, {1,2,3}

print("Values:\n\tsetA = {}\n\tsetB = {}".format(setA, setB))

# Intersection (&) between them
print("\n(setA & setB)= ", setA & setB)

# Union using (|) between them
print("(setA | setB)= ", setA | setB)
```

###**Subset and Superset**

-	Check if setB is a subset of setA using setB.issubset(setA).
-	Check if setA is a superset of setB using setA.issuperset(setB).
-	Print the results.


```python
# Given variable
setA, setB = {1, 2, 3, 4}, {1,2,3}

print("Original Sets:\n\tsetA = {}\n\tsetB = {}".format(setA, setB))
# Checking different condition using .issubset() and .issuperset()
print("\nIs setB is a subset of setA?\n= ", setB.issubset(setA))
print("Is setA is a superset of setB?\n= ", setA.issuperset(setB))
```
